Importing data 

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib Inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

Reading the data

```python
df = pd.read_csv(r"C:\Users\Krishna\Desktop\SQL Practice\Protfolio projects\movies.csv")
```

Lookin at the data 

```python
df
```

Looking at the nulls

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

Removing the incomeplete data

```python
df2 = df.dropna()
```

```python
df2.head(20)
```

Looking at the data types

```python
df2.dtypes
```

Changing data types of columns 

```python
df2['budget'] = df2['budget'].astype('int64')

df2['gross'] = df2['gross'].astype('int64')

df2['votes'] = df2['votes'].astype('int64')
```

Fixing the year released data 

```python
df2['yearfixed'] = df2['released'].astype('string').str.split(',').str[-1].astype('string').str[:5]
```

```python
df2.sort_values(by = 'gross',ascending = False)
```

Removing any duplicates

```python
df2 = df2.drop_duplicates()
```

```python
df2
```

Building scatter plot with budget vs gross

```python
plt.scatter(x = df['budget'],y = df['gross'] )

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show
```

Ploting budget vs gross using seaborn

```python
sns.regplot(x = 'budget', y = 'gross', data = df , scatter_kws = {'color':'red'},line_kws = {'color':'blue'} )
```

Loooking at correlation

```python
df2.corr(numeric_only = True)
```

```python
correlation_matrix = df2.corr(numeric_only = True , method = 'pearson')

sns.heatmap(correlation_matrix , annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Fatures')

plt.show()
```

```python
df_numerized = df2

for colname in df_numerized.columns:
    if(df_numerized[colname].dtype == 'object'):
      df_numerized[colname] = df_numerized[colname].astype('category')
      df_numerized[colname] = df_numerized[colname].cat.codes
```

```python
df_numerized 
```

```python
correlation_matrix = df_numerized .corr(numeric_only = True , method = 'pearson')

sns.heatmap(correlation_matrix , annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Fatures')

plt.show()
```
